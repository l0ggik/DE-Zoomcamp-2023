```python
df_1 = pd.read_csv('green_tripdata_2019-01.csv.gz')
df_1 = df_1[df_1['ehail_fee'].notna()]
df_2 = pd.read_csv('taxi+_zone_lookup.csv')
# df_2.rename(columns={'LocationID': })
```

```python
joined_df = pd.merge(df_1, df_2, left_on='PULocationID', right_on='LocationID')
```

```python
filtered_df = joined_df.loc[joined_df['Zone'] == 'Astoria']
result = filtered_df.loc[filtered_df['tip_amount'] == filtered_df['tip_amount'].max()]
result
```
